Este notebook realiza un análisis del conjunto de datos Breast Cancer de scikit-learn, incluyendo la carga de datos, exploración, preparación y entrenamiento de un modelo SVM.

### Carga de datos
Esta celda importa el módulo `datasets` de `sklearn` y carga el conjunto de datos de cáncer de mama en una variable llamada `datos`. El argumento `as_frame=True` asegura que los datos se carguen como un DataFrame de Pandas.

```python
from sklearn import datasets

datos = datasets.load_breast_cancer(as_frame=True)
```

### División de datos en conjuntos de entrenamiento y prueba
Esta celda importa `train_test_split` para dividir el conjunto de datos en conjuntos de entrenamiento y prueba. `X` contiene las características y `y` contiene la variable objetivo. Los datos se dividen con un 70% para el entrenamiento (`X_ent`, `y_ent`) y un 30% para la prueba (`X_pru`, `y_pru`).

```python
from sklearn.model_selection import train_test_split

X = datos.data
y = datos.target

X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size = .3)
```

### Entrenamiento y evaluación de modelos SVM (con un error común)
Esta celda importa SVM y métricas, luego itera a través de diferentes tipos de kernel ("linear", "rbf", "sigmoid") y valores de gamma para entrenar y evaluar modelos SVM. Para cada combinación, imprime las puntuaciones de exactitud y precisión.

**Nota:** En el código original, `modelo = svm.SVC()` se inicializa dentro del bucle sin pasar `kernel` y `gamma`. Esto significa que todos los modelos se entrenan con los valores predeterminados de `SVC` (kernel='rbf', gamma='scale'), independientemente de los valores de `kernel` y `gamma` en los bucles. El resultado de las impresiones mostrará los mismos valores para todos los kernels y gammas, porque el modelo no se está reconfigurando correctamente en cada iteración.

```python
from sklearn import svm
from sklearn import metrics

kernels = ["linear", "rbf", "sigmoid"]
gammas = [1, .01, .0001, .00001]

for kernel in kernels:
    for gamma in gammas:

      modelo = svm.SVC()
      modelo.fit(X_ent, y_ent)
      predicciones = modelo.predict(X_pru)

      print(f"{kernel}: Gamma: {gamma}: Exactitud:", metrics.accuracy_score(y_pru, predicciones))
      print(f"{kernel}: Gamma: {gamma}: Precision:", metrics.precision_score(y_pru, predicciones))
```

### Entrenamiento y evaluación de un modelo SVM lineal
Esta celda inicializa un modelo SVM con un kernel "linear", lo entrena con los datos de entrenamiento y luego realiza predicciones sobre los datos de prueba. Luego imprime las puntuaciones de exactitud y precisión para este modelo lineal específico.

```python
modelo = svm.SVC(kernel="linear")
modelo.fit(X_ent, y_ent)
predicciones = modelo.predict(X_pru)
print(f"{kernel}: Gamma: {gamma}: Exactitud:", metrics.accuracy_score(y_pru, predicciones))
print(f"{kernel}: Gamma: {gamma}: Precision:", metrics.precision_score(y_pru, predicciones))
```

### Informe de clasificación
Esta celda importa `classification_report` de `sklearn.metrics` e imprime un informe detallado del rendimiento de la clasificación, incluyendo precisión, recall, f1-score y soporte para cada clase.

```python
from sklearn.metrics import classification_report

print(classification_report(y_pru, predicciones))
```

### Matriz de confusión
Esta celda importa `confusion_matrix` y `pandas`, luego calcula y muestra la matriz de confusión para las predicciones del modelo. La matriz de confusión proporciona un desglose de las clasificaciones correctas e incorrectas por clase.

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

pd.DataFrame(confusion_matrix(y_pru, predicciones))
```
